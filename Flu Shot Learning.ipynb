## Categorical data

```python
c_imputed_feature = imputed_feature.copy()
le = LabelEncoder()
```

```python
c_imputed_feature['age_group'] = le.fit_transform(c_imputed_feature['age_group'])
c_imputed_feature['education'] = le.fit_transform(c_imputed_feature['education'])
c_imputed_feature['race'] = le.fit_transform(c_imputed_feature['race'])
c_imputed_feature['sex'] = le.fit_transform(c_imputed_feature['sex'])
c_imputed_feature['income_poverty'] = le.fit_transform(c_imputed_feature['income_poverty'])
c_imputed_feature['marital_status'] = le.fit_transform(c_imputed_feature['marital_status'])
c_imputed_feature['rent_or_own'] = le.fit_transform(c_imputed_feature['rent_or_own'])
c_imputed_feature['employment_status'] = le.fit_transform(c_imputed_feature['employment_status'])
c_imputed_feature['hhs_geo_region'] = le.fit_transform(c_imputed_feature['hhs_geo_region'])
c_imputed_feature['employment_industry'] = le.fit_transform(c_imputed_feature['employment_industry'])
c_imputed_feature['employment_occupation'] = le.fit_transform(c_imputed_feature['employment_occupation'])
c_imputed_feature['census_msa'] = le.fit_transform(c_imputed_feature['census_msa'])
```

```python
c_imputed_test_feature = imputed_test_feature.copy()

c_imputed_test_feature['age_group'] = le.fit_transform(c_imputed_test_feature['age_group'])
c_imputed_test_feature['education'] = le.fit_transform(c_imputed_test_feature['education'])
c_imputed_test_feature['race'] = le.fit_transform(c_imputed_test_feature['race'])
c_imputed_test_feature['sex'] = le.fit_transform(c_imputed_test_feature['sex'])
c_imputed_test_feature['income_poverty'] = le.fit_transform(c_imputed_test_feature['income_poverty'])
c_imputed_test_feature['marital_status'] = le.fit_transform(c_imputed_test_feature['marital_status'])
c_imputed_test_feature['rent_or_own'] = le.fit_transform(c_imputed_test_feature['rent_or_own'])
c_imputed_test_feature['employment_status'] = le.fit_transform(c_imputed_test_feature['employment_status'])
c_imputed_test_feature['hhs_geo_region'] = le.fit_transform(c_imputed_test_feature['hhs_geo_region'])
c_imputed_test_feature['employment_industry'] = le.fit_transform(c_imputed_test_feature['employment_industry'])
c_imputed_test_feature['employment_occupation'] = le.fit_transform(c_imputed_test_feature['employment_occupation'])
c_imputed_test_feature['census_msa'] = le.fit_transform(c_imputed_test_feature['census_msa'])
```

## Modeling

```python
RandomSeed = 6

X_train, X_test, y_train, y_test = train_test_split(
    c_imputed_feature,
    label, 
    test_size =0.35,
    stratify = label,
    shuffle= True,
    random_state = RandomSeed
)
```

### Logistic Model

```python
LogisticRegression_model = MultiOutputClassifier(estimator=LogisticRegression(solver = "liblinear"))
```

```python
LogisticRegression_model.fit(X_train, y_train)
```

```python
pred_base = LogisticRegression_model.predict_proba(X_test)
pred_base
```

```python
pred_base = pd.DataFrame({"h1n1_vaccine": pred_base[0][:, 1],"seasonal_vaccine": pred_base[1][:, 1],},index = y_test.index)
pred_base.head()
```

```python
roc_auc_score(y_test, pred_base)
```

```python
parameters = {'C': [1,5,50]}


clf_LogisticRegression= GridSearchCV(LogisticRegression(solver = "liblinear"), parameters, cv=8)
clf_LogisticRegression.fit(X_train, y_train["h1n1_vaccine"])

print("For h1n1_vaccine: ",clf_LogisticRegression.best_params_,clf_LogisticRegression.best_score_)

```

```python
pred_best_h = clf_LogisticRegression.predict_proba(X_test)
pred_best_h
```

```python
clf_LogisticRegression.fit(X_train, y_train["seasonal_vaccine"])

print("For seasonal_vaccine: ",clf_LogisticRegression.best_params_,clf_LogisticRegression.best_score_)
```

```python
pred_best_s = clf_LogisticRegression.predict_proba(X_test)
pred_best_s
```

```python
pred_best = pd.DataFrame({"h1n1_vaccine": pred_best_h[:, 1],"seasonal_vaccine": pred_best_s[:, 1],},index = y_test.index)
pred_best.head()
```
